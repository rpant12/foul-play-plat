```python
def strip_header_if_present(data):
    if len(data) > 524288:
        print("Detected extra data. Trimming to 512KB.")
        return data[-524288:]
    return data

def read_name_bytes_from_slot(slot, label):
    name_data = slot[0x34:0x34 + 8]
    print(f"{label} trainer name bytes:", name_data.hex(" "))
    return name_data

def decode_pokemon_text(byte_data):
    return ''.join(POKEMON_CHARSET.get(b, '?') for b in byte_data).rstrip('\0')

def get_save_slots(block, block_label):
    slot1 = block[0x0008:0x20008]
    slot2 = block[0x20008:0x40008]

    id1 = int.from_bytes(slot1[0xFFFC:0x10000], "little")
    id2 = int.from_bytes(slot2[0xFFFC:0x10000], "little")

    print(f"{block_label} - Slot1 ID: {id1}, Slot2 ID: {id2}")

    name1 = decode_pokemon_text(read_name_bytes_from_slot(slot1, f"{block_label} Slot1"))
    name2 = decode_pokemon_text(read_name_bytes_from_slot(slot2, f"{block_label} Slot2"))

    print(f"{block_label} Slot1 Trainer Name: {name1}")
    print(f"{block_label} Slot2 Trainer Name: {name2}")

    return (slot1, id1, name1), (slot2, id2, name2)



POKEMON_CHARSET = {
    0x00: '\0',
    0x01: 'A', 0x02: 'B', 0x03: 'C', 0x04: 'D', 0x05: 'E', 0x06: 'F', 0x07: 'G',
    0x08: 'H', 0x09: 'I', 0x0A: 'J', 0x0B: 'K', 0x0C: 'L', 0x0D: 'M', 0x0E: 'N',
    0x0F: 'O', 0x10: 'P', 0x11: 'Q', 0x12: 'R', 0x13: 'S', 0x14: 'T', 0x15: 'U',
    0x16: 'V', 0x17: 'W', 0x18: 'X', 0x19: 'Y', 0x1A: 'Z',
    0x1B: 'a', 0x1C: 'b', 0x1D: 'c', 0x1E: 'd', 0x1F: 'e', 0x20: 'f', 0x21: 'g',
    0x22: 'h', 0x23: 'i', 0x24: 'j', 0x25: 'k', 0x26: 'l', 0x27: 'm', 0x28: 'n',
    0x29: 'o', 0x2A: 'p', 0x2B: 'q', 0x2C: 'r', 0x2D: 's', 0x2E: 't', 0x2F: 'u',
    0x30: 'v', 0x31: 'w', 0x32: 'x', 0x33: 'y', 0x34: 'z',
    0x35: '0', 0x36: '1', 0x37: '2', 0x38: '3', 0x39: '4', 0x3A: '5',
    0x3B: '6', 0x3C: '7', 0x3D: '8', 0x3E: '9',
    0xFF: '\0'
}

def read_platinum_save(file_path):
    with open(file_path, "rb") as f:
        raw_data = f.read()

    data = strip_header_if_present(raw_data)

    block1 = data[0x00000:0x40000]
    block2 = data[0x40000:0x80000]

    print("Checking Block 1:")
    b1s1, b1s1_id, b1s1_name = get_save_slots(block1, "Block 1")
    print("\nChecking Block 2:")
    b2s1, b2s1_id, b2s1_name = get_save_slots(block2, "Block 2")

    # Determine best slot
    all_slots = [
        ("Block 1 Slot 1", b1s1_id, b1s1_name),
        ("Block 1 Slot 2", b1s1_id, b1s1_name),
        ("Block 2 Slot 1", b2s1_id, b2s1_name),
        ("Block 2 Slot 2", b2s1_id, b2s1_name),
    ]
    best = max(all_slots, key=lambda x: x[1])
    print(f"\n✅ Most recent save is in {best[0]} with Trainer Name: {best[2]}")

read_platinum_save("PokemonPlatinum.sav")
```

```python
def strip_header_if_present(data):
    if len(data) > 524288:
        print("Detected extra data. Trimming to 512KB.")
        return data[-524288:]
    return data

def get_active_block(data):
    block1 = data[0x00000:0x40000]
    block2 = data[0x40000:0x80000]

    id1 = int.from_bytes(block1[0xFFFC:0x10000], "little")
    id2 = int.from_bytes(block2[0xFFFC:0x10000], "little")

    print(f"Block1 Save ID: {id1}, Block2 Save ID: {id2}")
    return block1 if id1 >= id2 else block2

def get_active_slot(block):
    slot1 = block[0x0008:0x20008]
    slot2 = block[0x20008:0x40008]

    id1 = int.from_bytes(slot1[0xFFFC:0x10000], "little")
    id2 = int.from_bytes(slot2[0xFFFC:0x10000], "little")

    print(f"Slot1 Save ID: {id1}, Slot2 Save ID: {id2}")
    return slot1 if id1 >= id2 else slot2

def read_trainer_name(slot):
    name_offset = 0x34
    name_data = slot[name_offset : name_offset + 16]  # 8 UTF-16LE characters
    print("Raw trainer name bytes:", name_data.hex(" "))
    try:
        return name_data.decode("utf-16le").rstrip('\x00')
    except UnicodeDecodeError:
        return "Decode Error"

def read_platinum_save(file_path):
    with open(file_path, "rb") as f:
        raw_data = f.read()

    data = strip_header_if_present(raw_data)
    block = get_active_block(data)
    slot = get_active_slot(block)
    trainer_name = read_trainer_name(slot)

    print("Trainer Name:", trainer_name)

read_platinum_save("testsave.sav")
```
